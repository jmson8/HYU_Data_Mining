### login_logs
- 로그인 방법

```python
login_logs
```

```python
login_type = pd.DataFrame(login_logs.groupby('user_id')['login_method'].value_counts().groupby(level=0).idxmax().apply(lambda x: x[1]))
login_type.rename(columns={'count': 'login_method'}, inplace=True)
login_type
```

### subscription_logs
- 가입 기간, 구독 기간, 지불 금액, 지불 방법, 이탈 여부

```python
subscription_logs
```

```python
user_info = users.loc[:,['user_id','created_at']]
subscription_logs = pd.merge(subscription_logs,user_info,left_on='user_id',right_on='user_id')

subscription_logs['subscription_end'] = pd.to_datetime(subscription_logs['subscription_end'])
subscription_logs['subscription_start'] = pd.to_datetime(subscription_logs['subscription_start'])
subscription_logs['created_at'] = pd.to_datetime(subscription_logs['created_at'])

reference_date = subscription_logs['subscription_end'].max()

def calculate_months(start, end):
    return (end.year - start.year) * 12 + (end.month - start.month)

subscription_logs['signup_duration_months'] = subscription_logs['created_at'].apply(lambda x: calculate_months(x, reference_date))
subscription_logs['subscription_duration_months'] = subscription_logs.apply(lambda x: calculate_months(x['subscription_start'], x['subscription_end']), axis=1)
subscription_logs['active_classification'] = subscription_logs['subscription_end'].apply(
    lambda x: True if x.year == 2024 and x.month == 10 else False
)

subscription_info = subscription_logs.groupby('user_id').agg({
    'signup_duration_months': 'max', 'subscription_duration_months': 'sum', 'payment_amount': 'sum','payment_method': lambda x: x.value_counts().idxmax(), 
    'active_classification': 'max'
}).reset_index()

subscription_info['churn'] = ~subscription_info['active_classification']
subscription_info = subscription_info.drop(columns=['active_classification'])
subscription_info.columns = ['user_id', 'signup_duration', 'subscription_duration', 'payment_amount', 'payment_method', 'churn']

subscription_info
```

### view_logs
- 콘텐츠 시청 횟수, 추천 콘텐츠 시청 횟수, 최다 시청 카테고리, 시청 시간

```python
view_logs
```

```python
contents = contents.iloc[:,:2]
view_logs = pd.merge(view_logs, contents, left_on='content_id', right_on='content_id')

view_info = view_logs.groupby('user_id').agg({
    'watched_duration': 'sum', 'content_id': 'count','recommended': 'sum', 'category': lambda x: x.value_counts().idxmax() 
}).reset_index()

view_info['total_watched_duration_hours'] = view_info['watched_duration'] / 3600
view_info = view_info.drop(columns=['watched_duration'])
view_info.columns = ['user_id','content_view', 'recommended_view', 'category','watched_duration']

view_info
```

### interaction_logs
- 북마크 횟수, 좋아요 횟수, 싫어요 횟수, 평가 횟수, 총 상호작용 횟수

```python
interaction_logs
```

```python
interaction_type = pd.get_dummies(interaction_logs['interaction_type'])
interaction_logs = pd.concat([interaction_logs, interaction_type], axis=1)

interactions = interaction_logs.groupby('user_id').agg({
    'Bookmark': 'sum', 'Like': 'sum', 'Dislike': 'sum', 'Rating': 'sum', 'log_id': 'count'  
}).reset_index()
interactions.columns = ['user_id', 'bookmark_count', 'like_count', 'dislike_count', 'rating_count', 'total_interactions']

interactions
```

### 데이터 병합

```python
merged_df = pd.merge(users, login_type, on='user_id', how='inner')
merged_df = pd.merge(merged_df, subscription_info, on='user_id', how='inner')
merged_df = pd.merge(merged_df, view_info, on='user_id', how='outer')
merged_df = pd.merge(merged_df, interactions, on='user_id', how='outer')
```
